Importing all important libraries and Modules:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the Data set:

```python
df=pd.read_csv('/content/Loan.csv')
print('Rows & columns',df.shape)
df.head()
```

Describing data shape and other features:

```python

df.describe()
```

DataPerprocessing:


```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df=df.drop('ApplicationDate',axis=1)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
l_en=LabelEncoder()
df['EmploymentStatus']=l_en.fit_transform(df['EmploymentStatus'])
df['EducationLevel']=l_en.fit_transform(df['EducationLevel'])
df['MaritalStatus']=l_en.fit_transform(df['MaritalStatus'])
df['HomeOwnershipStatus']=l_en.fit_transform(df['HomeOwnershipStatus'])
df['LoanPurpose']=l_en.fit_transform(df['LoanPurpose'])
df.head()
```

Selecting Features:

```python
X=df.drop('LoanApproved',axis=1)
y=df['LoanApproved']
```

Data Splitting:

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=42)
```

Model selection and its training:


```python
model2=DecisionTreeClassifier()
model2.fit(X_train,y_train)
```

testing by testing data:

```python
y_predict=model2.predict(X_test)
```

Evaluation of model:

```python
print('Accuracy score:',accuracy_score(y_test,y_predict))
print('Confusion Matrix:\n',confusion_matrix(y_test,y_predict))
print('Classifiction Report:\n',classification_report(y_test,y_predict))
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_test,y_predict),fmt='d',annot=True,cmap='Oranges',cbar=False)
plt.xlabel('Predicted values')
plt.ylabel('Actual Values')
plt.title('Confusion Matrix')
```
